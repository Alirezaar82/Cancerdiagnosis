```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
data = pd.read_csv('cancer.csv')
```

```python
data.head(9)
```

```python
data.info()
```

```python
data.describe()
```

```python
data = data.drop(['id','Unnamed: 32'],axis=1)
data.diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis ]
```

```python
data.head(100)
```

```python
y_data = data.diagnosis
x_data = data.drop('diagnosis',axis=1)
# Normalize data
x_data = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
x_data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.15, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.max_iter = 1000000
model.fit(x_train, y_train)
```

```python
y_per = model.predict(x_test)
y_per
```

```python
y_test
```

```python
from sklearn import metrics


print("Accuracy:",metrics.accuracy_score(y_test, y_per) * 100 )
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy = accuracy_score(y_test, y_per)
cm = confusion_matrix(y_test, y_per)
report = classification_report(y_test, y_per)

# Print the results
print(f'Accuracy: {accuracy:.4f}')
print('Confusion Matrix:')
print(cm)
print('Classification Report:')
print(report)
```
